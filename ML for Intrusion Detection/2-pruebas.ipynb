```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Cargar datos
df = pd.read_csv('cybersecurity_intrusion_data.csv')

# Hacer una copia del DataFrame
df_copy = df.copy()

# Graficar histograma de 'network_packet_size'
plt.figure(figsize=(12, 8))
sns.histplot(df['network_packet_size'], bins=30, kde=True, color='#1f77b4')
plt.title("Distribucin del Tamao de los Paquetes", fontsize=18)
plt.xlabel("Tamao del Paquete (bytes)", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.tick_params(labelsize=12)
plt.tight_layout(pad=3)
plt.show()
```
